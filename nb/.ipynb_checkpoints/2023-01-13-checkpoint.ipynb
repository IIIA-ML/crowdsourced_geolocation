## 1.2 Recover sigmas

```python
t = 10000
k = 10
points = uniform.rvs(size=t)

t_A, w_A, ann = ann_set.random_annotation(k, points)
```

```python
def f(t, kappa, x, y):
    r = len(x)
    tr = np.repeat(t, r)
    tr.shape=(len(t), r)
    #spacing = x[1] - x[0]
    #print(d)
    d = (tr - x) * (tr - x)
    ed = np.exp(-kappa * d)
    s = np.sum(ed, axis=1)
    #print(s)
    res = np.dot(ed, y)/s
    #print(res)
    return res
```

```python
gp = cmd.CmdStanModel(stan_file='gp-learn-variances-ma.stan')
```

```python
a = len(ann)
l = 25
d = {"w":w,
     "a":a,
     "t":t,
     "t_A":t_A + 1,
     "w_A":w_A + 1,
     "ann":ann,
     "l":l
    }
inits = {"y_grid":np.ones((w,l))*0.1}
```

```python
s = gp.optimize(data=d, inits=inits, show_console=True, iter=1000000, algorithm='lbfgs', tol_rel_grad=10000.)
```

```python
gridpoints = np.arange(l) / (l - 1.)
```

```python
kappa = s.stan_variable("kappa")
y_grid = s.stan_variable("y_grid")
```

```python
for i, ann in enumerate(ann_set.annotators):
    sigmas = ann.sigma(allx)
    sigmas_learned = f(allx, kappa, gridpoints, y_grid[i])
    print("Annotator ",i,":", np.sqrt(np.mean(sigmas*sigmas)), np.sqrt(np.mean(sigmas_learned*sigmas_learned)))
    plt.plot(allx, sigmas, "b")
    plt.plot(allx, sigmas_learned, "r")
    plt.show()
#plt.legend(np.arange(w))
```

## 2.2 Strategies (should be on one_d.py module)

```python
class FoolStrategy: #We ask to M annotators randomly and compute the mean of the reported points
    
    def get_consensus(self, annotator_population, truth_points, max_annotator_pool):
        
        result = np.zeros((len(truth_points), max_annotator_pool))
        t_A, w_A, ann = ann_set.random_annotation(max_annotator_pool, truth_points)
        for p in range(len(truth_points)):
            result[p] = ann[t_A==p]
        
        consensus = np.mean(result, axis=1)
        return consensus
        
        
    def mean_error_per_sample_size(self, annotator_population, truth_points, max_annotator_pool):

        mean_error = []
        for sample_size in np.arange(2, max_annotator_pool):
            fool_reps = FoolStrategy().get_consensus(annotator_population, truth_points, sample_size)
            mean_error.append(np.mean(abs(truth_points - fool_reps)))

        return mean_error
    
    
    
class SmartStrategy: #We ask to M annotators randomly and compute the mean of their reports taking into account their sigmas
    
    def get_consensus(self, allx, annotator_population, truth_points, max_annotator_pool):
        
        consensus = []
        t_A, w_A, ann = ann_set.random_annotation(max_annotator_pool, truth_points)
        for i, p in enumerate(truth_points):
            consensus_dist = 1
            for annotator in range(max_annotator_pool):
                consensus_dist *= norm.pdf(allx, ann[t_A==i][annotator], annotator_population[w_A[t_A==i][annotator]].sigma(p))

            consensus.append(allx[np.argmax(consensus_dist)]) 
        return consensus
    
    
    def mean_error_per_sample_size(self, allx, annotator_population, truth_points, max_annotator_pool):

        mean_error = []
        for sample_size in np.arange(2, max_annotator_pool):
            smart_reps = SmartStrategy().get_consensus(allx, annotator_population, truth_points, sample_size)
            mean_error.append(np.mean(abs(truth_points - smart_reps)))

        return mean_error
        
```

## 2.3 Mean error in consensus depending on the number of annotators per point

```python
num_points = 1000
truth_points = uniform.rvs(0,1,num_points)
max_annotator_pool = 50
fool_mean_error = FoolStrategy().mean_error_per_sample_size(ann_set.annotators, truth_points, max_annotator_pool)
smart_mean_error = SmartStrategy().mean_error_per_sample_size(allx, ann_set.annotators, truth_points, max_annotator_pool)

plt.plot(np.arange(2, max_annotator_pool), fool_mean_error, 'bo', label = 'Fool')
plt.plot(np.arange(2, max_annotator_pool), smart_mean_error, 'go', label = 'Smart')
plt.title("Mean error between real and reported points")
plt.xlabel("Num annotators")
plt.xticks(np.arange(2, max_annotator_pool, 3))
plt.ylabel("Mean error")
plt.legend()
plt.show()
```

```python
class HillClimbing:
    
    def get_consensus(self, annotator_population, truth_points):
        
        consensus = []     
        for i,p in enumerate(truth_points):
            # Primero ordenamos los anotadores segn su sigma media:
            mean_sigmas = []
            for annotator in range(len(annotator_population.annotators)):
                mean_sigmas.append(np.mean(annotator_population.annotators[annotator].sigma(allx)))

            best_annotators = annotator_population.annotators[np.array(mean_sigmas).argsort()]

            # Hacemos que el mejor anotador anote el punto dos veces para establecer un intervalo:
            lims = np.sort(np.array([best_annotators[0].annotate(truth_points), 
                                 best_annotators[0].annotate(truth_points)]), axis=0)
            
            #Esta variable nos servir para detener las iteraciones:
            last_best = best_annotators[0]
            
            iterate = True
            while iterate == True: 
                print("LIMS:",lims)
                print("POINT:", p)
                
                # Volvemos a ordenar los anotadores segn su sigma media EN ESE INTERVALO:
                mean_sigmas = []
                for annotator in range(len(annotator_population.annotators)):
                    mean_sigmas.append(np.mean(annotator_population.annotators[annotator].sigma(np.arange(lims[0][i],
                                                                                                    lims[1][i],0.001))))

                print("MEAN SIGMAS IN INTERVAL:",mean_sigmas)
            
                best_annotators_aux = annotator_population.annotators[np.array(mean_sigmas).argsort()]
                print(best_annotators_aux)
                
                # Si el nuevo mejor anotador es el mismo de antes, ste anota y pasamos al siguiente punto.
                if best_annotators_aux[0] == last_best:
                    print("OK ANOTO")
                    consensus.append(best_annotators_aux[0].annotate(np.array([p])))
                    iterate = False
                    break
                # Si no, el nuevo mejor anotador vuelve a delimitar el intervalo:
                else:
                    lims = np.sort(np.array([best_annotators_aux[0].annotate(truth_points), 
                                 best_annotators_aux[0].annotate(truth_points)]), axis=0)
                    last_best = best_annotators_aux[0]
            
        return consensus
```
