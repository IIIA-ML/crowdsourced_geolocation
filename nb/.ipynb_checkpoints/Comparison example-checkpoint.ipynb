```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, uniform
import cmdstanpy as cmd
import seaborn as sn
from crowdgeoloc.one_d import sample_tokyo_latitudes, generate_annotators, annotate, SimpleNormalAnnotatorPopulation, RandomStrategy, RandomStrategyIntAverage, GreedyStrategyIntAverage

```

```python


def make_comparison(annotator_population, point_distribution, annotation_strategies, 
                    n_reps, n_annotators, n_points, max_annotator_pool):
    n_strategies = len(annotation_strategies)
    exp_result = np.zeros((n_reps, n_strategies, n_points, max_annotator_pool))
    points = np.zeros((n_reps, n_points))
    for r in range (n_reps):
        annotator_sample = annotator_population.sample(n_annotators)
        points[r,:] = uniform.rvs(size=n_points)
        for i, strategy in enumerate(annotation_strategies):
            exp_result[r,i,:,:] = strategy.location_curve(annotator_sample, points[r,:], max_annotator_pool)
    return points, exp_result

```

```python
n_reps = 100
n_annotators = 100
n_points = 100
max_annotator_pool = 10
annotator_population = SimpleNormalAnnotatorPopulation()
point_distribution = uniform()
annotation_strategies = [RandomStrategy(), RandomStrategyIntAverage(), GreedyStrategyIntAverage()]

p, e = make_comparison(annotator_population, point_distribution, annotation_strategies, n_reps, n_annotators, n_points, max_annotator_pool)
```

```python

def avg_distance_curve(p, e):
    p2 = p.copy()
    p2.shape=(100,1,100,1)
    p3 = np.repeat(p2,max_annotator_pool, axis=3)
    run_avg = np.average((e - p3)*(e-p3), axis =2)
    exp_avg = np.average(run_avg, axis=0)
    return exp_avg
```

```python
exp_avg = avg_distance_curve(p, e)
```

```python
for i in range(exp_avg.shape[0]):
    plt.plot(np.arange(10),exp_avg[i,:])
```
