# ROBUSTNESS

```python
class Annotator:
    def annotate(self, points: np.ndarray):
        pass

    def sigma(self, x: np.ndarray):
        pass
    
    
class SimpleOtherAnnotatorPopulation:
    def __init__(self, sigma_distribution=None):
        if sigma_distribution is None:
            sigma_distribution = beta(1, 10)
        self.sigma_distribution = sigma_distribution

    def sample(self, n_annotators):
        annotators = np.array([SimpleOtherAnnotator(sigma) for sigma in self.sigma_distribution.rvs(n_annotators)])
        return AnnotatorSet(annotators)
    
    
class AnnotatorSet:
    def __init__(self, annotators: np.ndarray):
        self.annotators = annotators

    def sample(self, k):
        selected_annotators_indexes = np.random.choice(len(self.annotators), size=k, replace=False)
        return self.annotators[selected_annotators_indexes]

    def random_annotation(self, k, points):
        n_points = len(points)
        t_A = np.zeros(n_points * k, dtype=int)
        w_A = np.zeros(n_points * k, dtype=int)
        annotations = np.zeros(n_points * k, dtype=float)
        selected_annotators_indexes = np.argsort(np.random.rand(n_points, len(self.annotators)), axis=1)[:, :k]
        total_annotations = 0
        for j, ann in enumerate(self.annotators):
            j_point_indices = np.argwhere(selected_annotators_indexes == j)[:, 0]
            j_points = points[j_point_indices]
            ann_points = ann.annotate(j_points)
            end_annotations = total_annotations + len(j_point_indices)
            t_A[total_annotations:end_annotations] = j_point_indices
            w_A[total_annotations:end_annotations] = j
            annotations[total_annotations:end_annotations] = ann_points
            total_annotations = end_annotations
        return t_A, w_A, annotations
```

```python
class SimpleOtherAnnotator(Annotator):
    def __init__(self, sigma):
        self._sigma = sigma

    def annotate(self, points: np.ndarray, community=0.1):
#         #Neightbour
#         home = self._sigma
#         annotations = []
#         for p in points:
#             if abs(p-home) < community:
#                 annotations.append(p)
#             else:
#                 annotations.append(np.random.uniform())

        #Approx
        zone = self._sigma
        annotations = []
        for p in points:
            if p + zone > 1:
                annotations.append(np.random.uniform(p-zone, 1))
            elif p - zone < 0:
                annotations.append(np.random.uniform(0, p+zone))
            else:
                annotations.append(np.random.uniform(p-zone, p+zone))
    
        return annotations
        
    def sigma(self, x):
        return self._sigma
```

# World creation (with real_sigmas = uniform(0,1))

```python
n_points = 1000
n_annotators = 100
redundancy = 4
```

```python
annotator_population = SimpleOtherAnnotatorPopulation(uniform())
point_distribution = uniform()
points = point_distribution.rvs(n_points)
ann_set = annotator_population.sample(n_annotators)
real_sigmas = np.array([x._sigma for x in ann_set.annotators])
annotations = ann_set.random_annotation(redundancy, points)
print("Num of annotations:", len(annotations[0]))
print("Real sigmas:", real_sigmas, "\n")
print(points)
```

# Sigmas recovery + (DataSet) Points smart consensus

## General strategy used: smart consensus

```python
def smart_consensus(annotations, real_points, recovered_sigmas, redundancy):
    
    x = np.arange(10000)/10000.
    t_A = annotations[0]

    consensus = []
    
    for i,p in enumerate(real_points):
        consensus_dist = 1
        for r in range(redundancy):
            corresponding_sigma_index = int(np.array(annotations)[:,t_A == i][1][r])
            corresponding_sigma = recovered_sigmas[corresponding_sigma_index]
            consensus_dist *= norm.pdf(x, np.array(annotations)[:,t_A == i][2][r], corresponding_sigma)
        
        consensus.append(x[np.argmax(consensus_dist)])
    
    mse = np.sum((real_points-consensus)**2)/len(real_points)
    
    return consensus, mse
```

## Direct estimation

### Sigmas recovery

```python
points_matrix = [[] for p in range(n_points)] #I group the annotations by the real points
for p in range(n_points):
    points_matrix[p] = np.array(annotations)[:,annotations[0]==p]
# print(points_matrix[0:5])


ann_deviation = dict((val, []) for val in range(n_annotators)) # I create a dictionary to store the difference between the real point and the annotation in each annotator
for p in range(n_points):
    mean_point = np.mean(points_matrix[p][2])
    for r in range(redundancy):
        ann_deviation[points_matrix[p][1][r]].append((mean_point - points_matrix[p][2][r])**2)

# print(len(ann_deviation[0])) # We can check that indeed we can get how many points an annotator has report
# print(len(np.array(annotations)[:,annotations[1]==0][0]))
```

```python
direct_sigmas = [] # Estimation of the sigmas by the std formula (with 1/(N-1))
for ann in range(n_annotators):
    direct_sigmas.append(np.sqrt(np.sum(ann_deviation[ann])/(len(ann_deviation[ann])-1)))
```

```python
plt.title(f"Direct estimation with {n_points} points and redundancy: {redundancy}")
plt.xlabel("Real sigmas")
plt.ylabel("Estimated sigmas")
plt.plot(real_sigmas, direct_sigmas,'yo', alpha=0.5)
plt.plot(np.linspace(0,1,1000), np.linspace(0,1,1000), 'g', alpha = 0.5)
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
```

```python
# Mean Squared Error:
direct_mse = np.sum((real_sigmas-direct_sigmas)**2)/len(real_sigmas)
print("MSE:", direct_mse)
```

### Points (from the DataSet) - smart consensus

## Iterative estimation

### Sigmas recovery

```python
# Should we gather together these functions into a single class?

def compute_avgs(my_avgs, annotations, sigmas):
    my_avgs.fill(0)
    weights = sigmas ** -2
    print(weights)
    denoms = np.zeros_like(my_avgs)
    # This should be numpyized
    point_indexes = annotations[0]
    annotator_indexes = annotations[1]
    values = annotations[2]
    n_annotations = len(point_indexes)
    for i in range(n_annotations):
        point_index = int(point_indexes[i])
        annotator_index = int(annotator_indexes[i])
        value = values[i]
        my_avgs[point_index, :] += (weights[annotator_index] * value)
        denoms[point_index, :] += weights[annotator_index]
#         my_avgs[point_index, annotator_index] -= (weights[annotator_index] * value)
#         denoms[point_index, annotator_index] -= weights[annotator_index]
        #if point_index == 0:
        #    print(my_avgs[0])
    my_avgs /= denoms
    return weights
    

def compute_sigmas(sigmas, annotations, avgs):
    n_points = avgs.shape[0]
    #print(n_points)
    sigmas.fill(0)
    #i=0
    point_indexes = annotations[0]
    annotator_indexes = annotations[1]
    values = annotations[2]
    n_annotations = len(point_indexes)
    counts = np.zeros_like(sigmas)
    for i in range(n_annotations):
        point_index = int(point_indexes[i])
        annotator_index = int(annotator_indexes[i])
        value = values[i]
        sq = ((avgs[point_index, annotator_index] - value) ** 2)
        #if i < 3:
            #print("Avg:",avgs[point_index, annotator_index])
            #print("Value:",value)
            #print("Sq:",sq)
        #    i += 1
        sigmas[annotator_index] += sq
        counts[annotator_index] += 1
        #print(sigmas)
    sigmas /= counts 
    np.sqrt(sigmas, out=sigmas)
    

def find_sigmas(annotations, n_points, n_annotators):
    my_avgs = np.zeros((n_points, n_annotators))
    sigmas = np.ones(n_annotators)
    i = 0
    sigmas_history = [sigmas.copy()]
    difference = 1
    while (i < 100) and (difference > 1e-16):
        #print(i, sigmas[:5])
        weights = compute_avgs(my_avgs, annotations, sigmas)
        #print(my_avgs[0:3])
        compute_sigmas(sigmas, annotations, my_avgs)
        difference = np.linalg.norm(sigmas_history[-1]-sigmas)
        if len(weights[weights[:] > 10**30]) != 0:
            break
        print("Change in this iteration:", difference)
        sigmas_history.append(sigmas.copy())
        i += 1
    return sigmas_history
```

## Conservative estimation

```python
# Should we gather together these functions into a single class?

def compute_avgs(my_avgs, annotations, sigmas):
    my_avgs.fill(0)
    weights = sigmas ** -2
    #print(weights)
    denoms = np.zeros_like(my_avgs)
    # This should be numpyized
    point_indexes = annotations[0]
    annotator_indexes = annotations[1]
    values = annotations[2]
    n_annotations = len(point_indexes)
    for i in range(n_annotations):
        point_index = int(point_indexes[i])
        annotator_index = int(annotator_indexes[i])
        value = values[i]
        my_avgs[point_index, :] += (weights[annotator_index] * value)
        denoms[point_index, :] += weights[annotator_index]
        my_avgs[point_index, annotator_index] -= (weights[annotator_index] * value)
        denoms[point_index, annotator_index] -= weights[annotator_index]
        #if point_index == 0:
        #    print(my_avgs[0])
    my_avgs /= denoms
    

def compute_sigmas(sigmas, annotations, avgs):
    n_points = avgs.shape[0]
    #print(n_points)
    sigmas.fill(0)
    #i=0
    point_indexes = annotations[0]
    annotator_indexes = annotations[1]
    values = annotations[2]
    n_annotations = len(point_indexes)
    counts = np.zeros_like(sigmas)
    for i in range(n_annotations):
        point_index = int(point_indexes[i])
        annotator_index = int(annotator_indexes[i])
        value = values[i]
        sq = ((avgs[point_index, annotator_index] - value) ** 2)
        #if i < 3:
            #print("Avg:",avgs[point_index, annotator_index])
            #print("Value:",value)
            #print("Sq:",sq)
        #    i += 1
        sigmas[annotator_index] += sq
        counts[annotator_index] += 1
        #print(sigmas)
    sigmas /= counts 
    np.sqrt(sigmas, out=sigmas)
    

def find_sigmas(annotations, n_points, n_annotators):
    my_avgs = np.zeros((n_points, n_annotators))
    sigmas = np.ones(n_annotators)
    i = 0
    sigmas_history = [sigmas.copy()]
    difference = 1
    while (i < 100) and (difference > 1e-16):
        #print(i, sigmas[:5])
        compute_avgs(my_avgs, annotations, sigmas)
        #print(my_avgs[0:3])
        compute_sigmas(sigmas, annotations, my_avgs)
        difference = np.linalg.norm(sigmas_history[-1]-sigmas)
        #print("Change in this iteration:", difference)
        sigmas_history.append(sigmas.copy())
        i += 1
    return sigmas_history
```
