## Implementation of Bayesian OneShot

````python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, uniform
import seaborn as sn
import pandas as pd


from crowdgeoloc.one_d import FunctionNormalAnnotatorPopulation, SimpleNormalAnnotatorPopulation
from crowdgeoloc.experiment import ActiveAnnotationContest, mean_location_norm_error, mean_sigma_norm_error
from crowdgeoloc.active import OneShotDirect, OneShotConservative, OneShotConservative2, OneShotIterative, \
    OneShotMean, KShot, OneShotBayesian, OneShotSpatialBayesian
````

> **NOTE**: If you want to run a *Stan* model directly (i.e., not via a `crowdgeoloc` class) within the notebook, use the above-imported `resource_file` function as in the example:
>
```ipython

```

````python
n_points = 1000
n_annotators = 200
redundancy = 4
````

````python
annotator_population = SimpleNormalAnnotatorPopulation(uniform())  # sigmas ~ uniform between 0 and 1
point_distribution = uniform()
points = point_distribution.rvs(n_points)
ann_set = annotator_population.sample(n_annotators)

exp = ActiveAnnotationContest(points, ann_set, max_total_annotations=n_points*redundancy)
````

````python
real_sigmas = np.array([x._sigma for x in ann_set.annotators])
plt.hist(real_sigmas)
plt.title("Real sigmas histogram")
plt.show()
````

````python
# from crowdgeoloc.active import random_annotation
# exp.reset()
# random_annotation(exp)
````

````python
exp.reset()
mean_consensus = OneShotMean().run(exp)
print("OneShotMean")
print("Average difference between real points and consensus:", np.mean(abs(points - mean_consensus['locations'])), '\n')

exp.reset()
direct_consensus = OneShotDirect().run(exp)
print("OneShotDirect")
print("Average difference between real points and consensus:", np.mean(abs(points - direct_consensus['locations'])),'\n')

exp.reset()
conservative2_consensus = OneShotConservative2().run(exp)
print("OneShotConservative2")
print("Average difference between real points and consensus:", np.mean(abs(points - conservative2_consensus['locations'])), '\n')

exp.reset()
kshot_consensus = KShot().run(exp)
print("KShot")
print("Average difference between real points and consensus:", np.mean(abs(points - kshot_consensus['locations'])), '\n')
````

````python
models = {"mean": OneShotMean(),
          "direct":OneShotDirect(), 
          #"iterative":OneShotIterative(), 
          "conservative2":OneShotConservative2(),
          "bayesian":OneShotBayesian(),
          #"conservative":OneShotConservative(),
          "10shot":KShot()
         }
metrics = {"mean_location_norm_error":mean_location_norm_error,
          "mean_sigma_norm_error":mean_sigma_norm_error}
````

````python
def compare(models, exp, metrics, repeats):
    results = []
    for model_name, m in models.items():
        print(model_name)
        result = {"model": model_name}
        for i in range(repeats):
            exp.reset()
            this_run = m.run(exp)
            result["iteration"] = i
            for metric_name, metric in metrics.items():
                result["metric"] = metric_name
                result["value"] = metric(exp, this_run)
                results.append(result.copy())
    return results
````

````python
results = compare(models, exp, metrics, 3)
df=pd.DataFrame(results)
df.head()
````

````python
sn.boxplot(data=df[df["metric"]=="mean_location_norm_error"], x="model", y="value")
plt.show()
sn.boxplot(data=df[df["metric"]=="mean_sigma_norm_error"], x="model", y="value")
````

````python
models = {"mean": OneShotMean(),
          "direct":OneShotDirect(), 
          #"iterative":OneShotIterative(), 
          "conservative2":OneShotConservative2(),
          "bayesian":OneShotBayesian(),
          #"conservative":OneShotConservative(),
          "10shot":KShot(),
          "spatialbayesian":OneShotSpatialBayesian()
         }
metrics = {"mean_location_norm_error":mean_location_norm_error}
          #"mean_sigma_norm_error":mean_sigma_norm_error}
````

````python
def compare(models, exp, metrics, repeats):
    results = []
    for model_name, m in models.items():
        print(model_name)
        result = {"model": model_name}
        for i in range(repeats):
            exp.reset()
            this_run = m.run(exp)
            result["iteration"] = i
            for metric_name, metric in metrics.items():
                result["metric"] = metric_name
                result["value"] = metric(exp, this_run)
                results.append(result.copy())
    return results
````
