```python
db = pd.read_csv("https://geographicdata.science/book/_downloads/7fb86b605af15b3c9cbd9bfcbead23e9/tokyo_clean.csv")
db.head()
```

```python
crs='epsg:4326'
geometry=[Point(xy) for xy in zip(db["longitude"], db["latitude"])]

gdf=gpd.GeoDataFrame(db,crs=crs, geometry=geometry)
gdf.head()
```

```python
kde = KernelDensity(bandwidth=0.001).fit(gdf[["longitude","latitude"]]) #entenc que fiteja els punts a una dist.

gdfs = sample(kde, crs=crs)
show_map(gdfs)
```

```python
#Little demo definitiva
num_annotators, annotators_errors = define_annotators(num_annotators=3, 
                                                      annotators_errors=[[0.05, 0.2], [0.6, 0.1], [0.01, 0.01]])

reported_points = multi_annotators_rep(gdfs, num_annotators, annotators_errors)


consensus_points = mean_consensus(reported_points)


print(best_accuracy_weight(annotators_errors))
print(accuracy_weight(annotators_errors))
```

```python
from scipy.stats import norm
rv = norm()

x1=1
x2=3
x3=4
sigma1=0.4
sigma2=0.2
sigma3=0.7
def logf(x):
    return norm.logpdf(x1, loc=x, scale=sigma1) + norm.logpdf(x2, loc=x, scale=sigma2)
#+ norm.logpdf(x3, loc=x, scale=sigma3)
def f(x):
    return np.exp(logf(x))

xs=np.arange(-1,10,0.01)
```
