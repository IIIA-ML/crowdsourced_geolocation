## Comparison between stan and iterative method with sigma_distribution = uniform()

### Iterative method:

```python
n_points = 1000
n_annotators = 40
redundancy=3
annotator_population = SimpleNormalAnnotatorPopulation(uniform())
point_distribution = uniform()
points = point_distribution.rvs(n_points)
ann_set = annotator_population.sample(n_annotators)
real_sigmas = np.array([x._sigma for x in ann_set.annotators])
annotations = ann_set.random_annotation(redundancy, points)
print(len(annotations[0]))
print("Real sigmas:", real_sigmas, "\n")
```

```python
# Should we gather together these functions into a single class?

def compute_avgs(my_avgs, annotations, sigmas):
    my_avgs.fill(0)
    weights = sigmas ** -2
    #print(weights)
    denoms = np.zeros_like(sigmas)
    sum_weights = np.sum(weights)
    denoms.fill(sum_weights)
    denoms -= weights
    #print(denoms)
    # This should be numpyized
    point_indexes = annotations[0]
    annotator_indexes = annotations[1]
    values = annotations[2]
    n_annotations = len(point_indexes)
    for i in range(n_annotations):
        point_index = int(point_indexes[i])
        annotator_index = int(annotator_indexes[i])
        value = values[i]
        my_avgs[point_index, :] += (weights[annotator_index] * value)
        my_avgs[point_index, annotator_index] -= (weights[annotator_index] * value)
        #if point_index == 0:
        #    print(my_avgs[0])
    my_avgs /= denoms[None,:]
    

def compute_sigmas(sigmas, annotations, avgs):
    n_points = avgs.shape[0]
    #print(n_points)
    sigmas.fill(0)
    #i=0
    point_indexes = annotations[0]
    annotator_indexes = annotations[1]
    values = annotations[2]
    n_annotations = len(point_indexes)
    for i in range(n_annotations):
        point_index = int(point_indexes[i])
        annotator_index = int(annotator_indexes[i])
        value = values[i]
        sq = ((avgs[point_index, annotator_index] - value) ** 2)
        #if i < 3:
            #print("Avg:",avgs[point_index, annotator_index])
            #print("Value:",value)
            #print("Sq:",sq)
        #    i += 1
        sigmas[annotator_index] += sq
        #print(sigmas)
    sigmas /= (n_points - 1)
    np.sqrt(sigmas, out=sigmas)
    

def find_sigmas(annotations, n_points, n_annotators):
    my_avgs = np.zeros((n_points, n_annotators))
    sigmas = np.ones(n_annotators)
    i = 0
    sigmas_history = [sigmas.copy()]
    difference = 1
    while (i < 100) and (difference > 1e-16):
        #print(i, sigmas[:5])
        compute_avgs(my_avgs, annotations, sigmas)
        #print(my_avgs[0:3])
        compute_sigmas(sigmas, annotations, my_avgs)
        difference = np.linalg.norm(sigmas_history[-1]-sigmas)
        #print("Change in this iteration:", difference)
        sigmas_history.append(sigmas.copy())
        i += 1
    return sigmas_history
```

```python
#print(points[:4])
#print(real_sigmas)
sigmas_history=find_sigmas(annotations, n_points, n_annotators)
```

```python
learned_sigmas = sigmas_history[-1]
print("Real sigmas:", real_sigmas, "\n")
print("Recovered sigmas", learned_sigmas, "\n")
print("Mean of the difference:", np.mean(np.abs(real_sigmas-learned_sigmas)))
```

```python

plt.plot(real_sigmas, learned_sigmas,'bo', alpha=0.5)
plt.plot(np.linspace(0,1,1000), np.linspace(0,1,1000), 'g', alpha = 0.5)
plt.xlim(0,1)
plt.ylim(0,1)
```

```python
# Mean Squared Error:
iterative_mse = np.sum((real_sigmas-learned_sigmas)**2)/len(real_sigmas)
print("MSE:", iterative_mse)
```

### Stan:

```python
cmd.install_cmdstan()
```

```python
model = cmd.CmdStanModel(stan_file='normal.2.stan')
```

```python
d = {"w":n_annotators,
     "a":len(annotations[0]),
     "t":n_points,
     "t_A":annotations[0] + 1,
     "w_A":annotations[1] + 1,
     "ann":annotations[2]
    }
inits = {"sigma": [np.ones(n_annotators)]}
```

```python
s = model.sample(data=d, inits=inits, show_console=True)
```

```python
rec_sigmas_sample = s.stan_variable("sigmas")
rec_sigmas = []
for annotator in range(n_annotators):
    rec_sigmas.append(np.median(rec_sigmas_sample[:,annotator]))  
```

```python
print("REAL SIGMAS: ", real_sigmas)
print("RECOVERED SIGMAS: ", np.round(rec_sigmas, 4))
print("Mean of the difference:", np.mean(np.abs(real_sigmas-rec_sigmas)))
```

```python

plt.plot(real_sigmas, rec_sigmas,'ro', alpha=0.5)
plt.plot(np.linspace(0,1,1000), np.linspace(0,1,1000), 'g', alpha = 0.5)
plt.xlim(0,1)
plt.ylim(0,1)
```

```python
# Mean Squared Error:
stan_mse = np.sum((real_sigmas-rec_sigmas)**2)/len(real_sigmas)
print("MSE:", stan_mse)
```

### Plot comparison

```python
plt.plot(real_sigmas, learned_sigmas, 'bo', label = f'Iterative - MSE: {round(iterative_mse,4)}')
plt.plot(real_sigmas, rec_sigmas, 'ro', label = f'Stan - MSE: {round(stan_mse,4)}', alpha = 0.5)
plt.plot(np.linspace(0,1,1000), np.linspace(0,1,1000), 'g')
plt.legend()
plt.show()
```
