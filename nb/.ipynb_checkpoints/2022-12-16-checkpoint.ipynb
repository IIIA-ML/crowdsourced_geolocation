```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

from crowdgeoloc.simple_module import *
```

```python
cmd.install_cmdstan()
```

```python
number_of_points = 2000

real_points = np.random.random_sample(number_of_points)
num_annotators = 25
map_parts = 1
annotators_sigmas = np.round(np.random.random_sample(num_annotators),3)

rep_points = annotators_reports(real_points, num_annotators, annotators_sigmas, map_parts)
```

```python
n = number_of_points
l = num_annotators

print(n,l,)

d = {"N":n,"L":l,"x":rep_points}
#inits = {"sigma":[[np.round(np.max(stds), 3) for i in range(num_annotators)]]}
inits = {"sigma": [np.ones(l)]}
```

```python
model = cmd.CmdStanModel(stan_file='normal_model.stan')
```

```python
s = model.sample(data=d, inits=inits, show_console=True)
```

```python
rec_sigmas_sample = s.stan_variable("sigma")
```

```python
rec_sigmas = []
for annotator in range(num_annotators):
    rec_sigmas.append(np.median(rec_sigmas_sample[:,annotator]))  
```

```python
print("REAL SIGMAS: ", annotators_sigmas)
print("RECOVERED SIGMAS: ", np.round(rec_sigmas, 4))
```

```python
slope, intercep, r_value, p_value, std_err = linregress(annotators_sigmas, rec_sigmas)

plt.xlim(0,1)
plt.ylim(0,1)
fig = sn.regplot(x=annotators_sigmas, y=rec_sigmas, line_kws = {'lw':1, 'color':'red'}, truncate=False)
fig.set(xlabel="REAL", ylabel="RECOVERED")
plt.text(0.7,0.1,"R^2 = {0}".format(np.round(r_value**2,4)))
plt.text(0.7,0.2,"y={0}x+{1}".format(np.round(slope,4), np.round(intercep,4)))
```

```python
num_annotators = 20
annotators_sigmas = np.round(np.random.random_sample(num_annotators),3)
num_annotators_per_group = 5

rep_points = reports_frompool(real_points, num_annotators, annotators_sigmas, num_annotators_per_group)
#print(rep_points[:5], "\n")

marked_annotators = mark_annotators_frompool(rep_points)
#print(marked_annotators[:5])
```

```python
n = number_of_points
l = num_annotators
marked = marked_annotators

pool_d = {"N":n,"L":l,"marked":marked,"x":rep_points}
#inits = {"sigma":[[np.round(np.max(stds), 3) for i in range(num_annotators)]]}
pool_inits = {"sigma": [np.ones(l)]}
```

```python
slope, intercep, r_value, p_value, std_err = linregress(annotators_sigmas, pool_rec_sigmas)

plt.xlim(0,1)
plt.ylim(0,1)
fig = sn.regplot(x=annotators_sigmas, y=pool_rec_sigmas, line_kws = {'lw':1, 'color':'red'}, truncate=False)
fig.set(xlabel="REAL", ylabel="RECOVERED")
plt.text(0.7,0.1,"R^2 = {0}".format(np.round(r_value**2,4)))
plt.text(0.7,0.2,"y={0}x+{1}".format(np.round(slope,4), np.round(intercep,4)))
plt.show()
```

```python
number_of_points = [1000]
data = ["Tokio"]
pool = ["No"]
num_annotators = [3, 5, 10]
```

```python
for number_of_points_type in number_of_points:
    
    for data_type in data:
        if data_type == "Tokio":
            real_points = read_Tokio_1D(number_of_points_type) #WATCHOUT: Should it be normalized?
        else:
            real_points = np.random.random_sample(number_of_points_type)
        
        for num_annotators_type in num_annotators:
            annotators_sigmas = np.round(np.random.random_sample(num_annotators_type),3)
            for pool_type in pool:
                if pool_type == "Yes":
                    num_annotators_per_group = int(np.ceil(num_annotators_type/2))
                    rep_points = reports_frompool(real_points, num_annotators_type, annotators_sigmas, num_annotators_per_group)
                    marked_annotators = mark_annotators_frompool(rep_points)

                else:
                    num_annotators_per_group = "Not in use"
                    rep_points = annotators_reports(real_points, num_annotators_type, annotators_sigmas, 1)
                    marked_annotators = np.ones(np.array(rep_points).shape)
                    
                print(
                    "Data:", data_type,"\n",
                    "Number of points:", number_of_points_type,"\n",
                    "Number of annotators:", num_annotators_type,"\n",
                    "Pool of annotators:", pool_type,"\n",
                    "Number of annotators per group:", num_annotators_per_group,"\n"
                      )
                
#                 print(
#                     "Real points:", real_points,"\n",
#                     "Reported points:", rep_points,"\n"
#                 )


                n = number_of_points_type
                l = num_annotators_type
                marked = marked_annotators
                pool_d = {"N":n,"L":l,"marked":marked,"x":rep_points}
                pool_inits = {"sigma": [np.ones(l)]}
                
                pool_s = pool_model.sample(data=pool_d, inits=pool_inits)
                pool_rec_sigmas_sample = pool_s.stan_variable("sigma")
                pool_rec_sigmas = []
                for annotator in range(num_annotators_type):
                    pool_rec_sigmas.append(np.median(pool_rec_sigmas_sample[:,annotator]))  

                print("REAL SIGMAS: ", annotators_sigmas)
                print("RECOVERED SIGMAS: ", np.round(pool_rec_sigmas, 4))
                
                slope, intercep, r_value, p_value, std_err = linregress(annotators_sigmas, pool_rec_sigmas)
                plt.xlim(0,1)
                plt.ylim(0,1)
                fig = sn.regplot(x=annotators_sigmas, y=pool_rec_sigmas, line_kws = {'lw':1, 'color':'red'}, truncate=False)
                fig.set(xlabel="REAL", ylabel="RECOVERED")
                plt.text(0.7,0.1,"R^2 = {0}".format(np.round(r_value**2,4)))
                plt.text(0.7,0.2,"y={0}x+{1}".format(np.round(slope,4), np.round(intercep,4)))
                plt.show()
```
